```python
import cv2
# Importing Necessary Libraries.
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

```

Load the Original Image(RGB image)

```python
img = cv2.imread("dog.jpg")
cv2_imshow(img)
```

Convert the RGB image to grayscale image

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_img)
```

Invert the Grayscale image

```python
invert_img = 225-gray_img
cv2_imshow(invert_img)
```

Smoothen the inverted image using gaussian blur technique

```python
img_smoothen = cv2.GaussianBlur(invert_img, (21,21), sigmaX=0, sigmaY=0)
cv2_imshow(img_smoothen)
```

Now create a pencil sketch image

```python
inverted_blurr = 255-img_smoothen
pencil_sketch = cv2.divide(gray_img, invert_img,inverted_blurr, scale = 256.0)
cv2_imshow(pencil_sketch)
```
